```python
import glob
import os
import copy
from PIL import Image
from PIL import ImageEnhance
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#variables
sharpeenhance=3
traindatapath="..\database\raw\train"
testdatapath="..\database\raw\test"

#this reads in the training data and output them
list=[]
listsharp=[]

#this reads in the testing data and output them
list2=[]
listsharp2=[]

sharpnessenhancenum=5
contrastenhancenum=1.2
brightnessenhancenum=2
colourenhancenum=3.5

```

```python
#read in the data(train)
list.clear()
for counter in range(0,57+1):
    newlist=[]  
    for k in glob.glob(r"..\database\raw\train\%s\*.png"%str(counter)):
        temp=Image.open(k)
        keep=temp.copy()
        newlist.append(keep)
        temp.close()
        
    list.append(newlist)
```

```python
#read in the data(test)
list2.clear()
for counter in range(0,57+1):
    newlist=[]  
    for k in glob.glob(r"..\database\raw\test\%s\*.png"%str(counter)):
        temp=Image.open(k)
        keep=temp.copy()
        newlist.append(keep)
        temp.close()
        
    list2.append(newlist)
```

```python
plt.imshow(list[2][0])
```

```python
listsharp.clear()
for i in range(len(list)):
    templist=[]
    for j in range(len(list[i])):
        copy=list[i][j].copy()
        en1=ImageEnhance.Sharpness(copy).enhance(sharpnessenhancenum)
        en2=ImageEnhance.Contrast(en1).enhance(contrastenhancenum)
        en3=ImageEnhance.Brightness(en2).enhance(brightnessenhancenum)
        en4=ImageEnhance.Color(en3).enhance(colourenhancenum)
        templist.append(en3)
    listsharp.append(templist)

        
       
```

```python
listsharp2.clear()
for i in range(len(list2)):
    templist=[]
    for j in range(len(list2[i])):
        copy=list2[i][j].copy()
        en1=ImageEnhance.Sharpness(copy).enhance(sharpnessenhancenum)
        en2=ImageEnhance.Contrast(en1).enhance(contrastenhancenum)
        en3=ImageEnhance.Brightness(en2).enhance(brightnessenhancenum)
        en4=ImageEnhance.Color(en3).enhance(colourenhancenum)
        templist.append(en3)
    listsharp2.append(templist)
```

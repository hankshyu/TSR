```python
import glob
import os
import copy
from PIL import Image
from PIL import Image
from PIL import ImageEnhance
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#variables
sharpeenhance=3
traindatapath="..\database\raw\train"
testdatapath="..\database\raw\test"
list=[]
listsharp=[]

list2=[]
listsharp2=[]
```

```python
#read in the data(train)
list.clear()
for counter in range(0,57+1):
    newlist=[]  
    for k in glob.glob(r"C:\Users\User\Desktop\TSR\database\raw\train\%s\*.png"%str(counter)):
        temp=Image.open(k)
        keep=temp.copy()
        newlist.append(keep)
        temp.close()
        
    list.append(newlist)
```

```python
#read in the data(test)
list2.clear()
for counter in range(0,57+1):
    newlist=[]  
    for k in glob.glob(r"C:\Users\User\Desktop\TSR\database\raw\test\%s\*.png"%str(counter)):
        temp=Image.open(k)
        keep=temp.copy()
        newlist.append(keep)
        temp.close()
        
    list2.append(newlist)
```

```python
plt.imshow(list[2][0])
```

```python
print(len(list2))
for i in list2:
    print(str(len(i))+"\t",end="")
```

```python
listsharp.clear()
for i in range(len(list)):
    templist=[]
    for j in range(len(list[i])):
        copy=list[i][j].copy()
        en1=ImageEnhance.Sharpness(copy).enhance(3)
        en2=ImageEnhance.Contrast(en1).enhance(3)
        templist.append(en2)
    listsharp.append(templist)

        
       
```

```python
listsharp2.clear()
for i in range(len(list2)):
    templist=[]
    for j in range(len(list2[i])):
        copy=list2[i][j].copy()
        en1=ImageEnhance.Sharpness(copy).enhance(3)
        en2=ImageEnhance.Contrast(en1).enhance(3)
        templist.append(en2)
    listsharp2.append(templist)
```

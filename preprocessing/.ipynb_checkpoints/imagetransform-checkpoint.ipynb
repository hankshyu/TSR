```python
from PIL import Image
from PIL import ImageEnhance
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
image1=Image.open("one.png")
image2=Image.open("two.png")
image3=Image.open("three.png")
```

```python
plt.imshow(image1)
```

```python
print(str(image1.size)+"\t"+str(image1.format)+"\t"+str(image1.mode))
print(str(image2.size)+"\t"+str(image2.format)+"\t"+str(image2.mode))
print(str(image3.size)+"\t"+str(image3.format)+"\t"+str(image3.mode))
```

```python
copy=image1.copy()
en1=ImageEnhance.Sharpness(copy).enhance(4)
plt.imshow(en1)
```

```python
copy=image1.copy()
en1=ImageEnhance.Contrast(copy).enhance(2)
plt.imshow(en1)
```
